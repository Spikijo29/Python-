```python
# Distibution of PlayeventType
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel (data_path)
df['PlayEventType'].value_counts().plot(kind='bar',
                                        title='Distribution of PlayEventType' 
)
plt.xlabel('PlayEventType')
plt.ylabel('Count')
plt.show()
```

```python
# Top 10 Most Viewed VideoTitle

import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel (data_path)
df['VideoTitle'].value_counts().head(10).plot(kind='bar',
                                        title='Top 10 Most Viewed VideoTitle' 
)
plt.xlabel('VideoTitle')
plt.ylabel('Count')
plt.show()
```

```python
#Distribution of viewership over time
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel (data_path)
if 'TotalTimeWatched' in df.columns:
    df['TotalTimeWatched'] = pd.to_datetime(df['TotalTimeWatched'])
    df.set_index('TotalTimeWatched').resample('D').size().plot(
        title='Viewership Over Time'
    )
    plt.xlabel('Date')
    plt.ylabel('Count')
    plt.show()

```
